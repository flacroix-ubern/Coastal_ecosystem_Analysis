```python
vars_marcats=pd.read_csv("tables/coastal_vars_observations.csv")

###Plot figure
#fig, ax =  plt.subplots(1,4,figsize=(12,5))
###get data
vars_marcats_index = vars_marcats["Index"]
index_regions = (vars_marcats_index.eq(23)) | (vars_marcats_index.eq(5))
iconic_regions_pd = vars_marcats[(vars_marcats["Index"].eq(43)) | (vars_marcats["Index"].eq(5)) | (vars_marcats["Index"].eq(39)) | (vars_marcats["Index"].eq(38)) | (vars_marcats["Index"].eq(31)) | (vars_marcats["Index"].eq(17)) | (vars_marcats["Index"].eq(2))]
iconic_regions_pd=iconic_regions_pd.reindex(index=[43, 5, 17,39,2,31,38])
iconic_regions_pd.loc[iconic_regions_pd["Index"]==2,"SST OBS"]=0.36
iconic_regions_pd.loc[iconic_regions_pd["Index"]==5,"SST OBS"]=0.78
iconic_regions_pd.loc[iconic_regions_pd["Index"]==43,"H+ OBS"]=1.36*1E-9
iconic_regions_pd["Oxygen OBS"]=np.flip([-0.12,-0.15,0.03,-0.11,-0.08,0.06,0.02])
print(iconic_regions_pd)

iconic_regions_pd["SST OBS"] = [0.52,0.12,1.43,1.64,0.09,0.56,0.62]
iconic_regions_pd["H+ OBS"] = [1.87,1.04,1.35,1.42,0.91,1.36,1.31]
iconic_regions_pd["NPP OBS"] = [36.4,25.7,38.2,29.2,14.5,27.3,21.5]
iconic_regions_pd["Oxygen OBS"] = [-0.04,0.03,-0.42,-0.31,-0.03,-0.16,0.04]
SST_in_situ = [0.21,0.06,2.0,1.2,1.04,0.81,1.6]
iconic_regions_pd["SST OBS"] = iconic_regions_pd["SST OBS"] + ( SST_in_situ - iconic_regions_pd["SST OBS"] ) / 2

uncertainty = np.array(SST_in_situ) - np.array([0.52,0.12,1.43,1.64,0.19,0.66,0.52])
###plot colors
sst_colors=["orange"]
ph_colors="darkkhaki"
npp_colors=["darkred","orange","darkred","orange","darkred","orange","darkred"][::-1]  # Flip
oxygen_colors=["orange","darkred","orange","darkred","orange","orange","orange","orange"][::-1]

####Bar plot
height=6
count=0
for row in iconic_regions_pd.iterrows():
    ax[0].plot([row[1]["SST OBS"],row[1]["SST OBS"]]-uncertainty[count]/2,[height-0.05,height+0.05],linewidth=3,linestyle="-",color="grey")
    ax[0].plot([row[1]["SST OBS"],row[1]["SST OBS"]]+uncertainty[count]/2,[height-0.05,height+0.05],linewidth=3,linestyle="-",color="grey")
    ax[0].plot([row[1]["SST OBS"]-uncertainty[count]/2,row[1]["SST OBS"]+uncertainty[count]/2],[height,height],linewidth=3,linestyle="-",color="grey")
    ax[1].plot([row[1]["H+ OBS"],row[1]["H+ OBS"]],[height-0.2,height+0.2],linewidth=4,color="grey")
    ax[2].plot([row[1]["NPP OBS"],row[1]["NPP OBS"]],[height-0.2,height+0.2],linewidth=4,color="grey")
    print(row[1]["Oxygen OBS"])
    ax[3].plot([row[1]["Oxygen OBS"],row[1]["Oxygen OBS"]],[height-0.2,height+0.2],linewidth=4,color="grey")
    height=height-1
    count=count+1


colors = {'anthr. CO$_{2}$':'darkkhaki', 'climate':'orange',"ant. nutrients":"darkred"}         
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
print(handles)
ax[3].legend(handles, labels,loc="lower right",frameon=False,bbox_to_anchor=(+1.7,0.35),fontsize=10)
fig.savefig("figures/top_regions_legendpart1.png",dpi=400,bbox_inches = "tight")
#ypos=np.arange(0,7)
#ax[0].barh(y=np.array(ypos),width=np.array(iconic_regions_pd["SST OBS"][::-1]),color=sst_colors,edgecolor='black')
#ax[1].barh(y=np.array(ypos),width=np.array(iconic_regions_pd["H+ OBS"][::-1])*1E9,color=ph_colors,edgecolor='black')
#ax[2].barh(y=np.array(ypos),width=np.array(iconic_regions_pd["NPP OBS"][::-1]),color=npp_colors,edgecolor='black')
#ax[3].barh(y=np.array(ypos),width=np.array(iconic_regions_pd["Oxygen OBS"][::-1]),color=oxygen_colors,edgecolor='black')
```
